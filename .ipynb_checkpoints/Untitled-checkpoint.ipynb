```python
def load_and_predict(tenor: str, model: str, df: pd.DataFrame):
    
    with open(f"streamlit/pages/models/state_false/{tenor}_models_{model.lower()}_bayes.pkl", "rb") as file:
        params_model = pickle.load(file)
    
    scaler = StandardScaler()
    scaler.mean_, scaler.scale_ = params_model[model.lower()]['scaler_train_mean'], params_model[model.lower()]['scaler_train_std']

    # transform the dataframe
    df_scaled = scaler.transform(df)
    
    # reformat the scaled data frame into X (input), y(target)
    X_scaled, y_scaled = reformat_to_arrays(df_scaled, n_steps = 15, n_steps_ahead = 4)
    
    # predict with the model
    
    model = params_model[model.lower()]['model']
    predictions = model.predict(X_scaled)
    predictions = scaler.inverse_transform(predictions)
    ground_truth = scaler.inverse_transform(y_scaled)
    
    return params_model, predictions, ground_truth
```

```python
def get_ground_truth_continuous(all_ground_truth):
    
    ground_truth_continuous = []

    # Add the first value of each 4-day array, except for the last one
    for i in range(len(all_ground_truth) - 1):
        ground_truth_continuous.append(all_ground_truth[i, 0])

    # Add the entire last 4-day array
    ground_truth_continuous.extend(all_ground_truth[-1])

    # Convert to numpy array (optional)
    ground_truth_continuous = np.array(ground_truth_continuous)

    return ground_truth_continuous
```

```python
# Read the data

# 2-year
df_2yr = pd.read_csv("data/raw/bond_yields - USGG2YR.csv", index_col=0)
df_2yr.index = pd.to_datetime(df_2yr.index) # infer_datetime_format=True)
df_2yr_H1_2020 = df_2yr[(df_2yr.index >= '2020-01-01') & (df_2yr.index <= '2020-06-30')]
df_2yr_H1_2023 = df_2yr[(df_2yr.index >= '2023-01-01') & (df_2yr.index <= '2023-06-30')]


#10-year
df_10yr = df_2yr = pd.read_csv("data/raw/bond_yields - USGG10YR.csv", index_col=0)
df_10yr.index = pd.to_datetime(df_10yr.index) # infer_datetime_format=True)
df_10yr_H1_2020 = df_10yr[(df_10yr.index >= '2020-01-01') & (df_10yr.index <= '2020-06-30')]
df_10yr_H1_2023 = df_10yr[(df_10yr.index >= '2023-01-01') & (df_10yr.index <= '2023-06-30')]
```

```python
dataset_="2yr"
params_rnn, rnn_predictions, ground_truth = load_and_predict(dataset_, 'rnn', df_2yr_H1_2023)
params_gru, gru_predictions, ground_truth = load_and_predict(dataset_, 'gru', df_2yr_H1_2023)
params_lstm, lstm_predictions, ground_truth = load_and_predict(dataset_, 'lstm', df_2yr_H1_2023)

# Convert ground truth into a single continuous array
ground_truth_continuous = []
for i in range(len(ground_truth) - 1):
    ground_truth_continuous.append(ground_truth[i, 0])
ground_truth_continuous.extend(ground_truth[-1])
ground_truth_continuous = np.array(ground_truth_continuous)

# Time axis for ground truth
time_axis_truth = np.arange(len(ground_truth_continuous))

# Initialize arrays for averaged predictions
time_series_length = len(rnn_predictions) + 3  # Include overlapping days
predicted_full_rnn = np.zeros(time_series_length)
predicted_full_gru = np.zeros(time_series_length)
predicted_full_lstm = np.zeros(time_series_length)

overlap_count = np.zeros(time_series_length)  # To average overlapping predictions

# Fill the arrays for predictions (for RNN, GRU, and LSTM)
for i, (rnn_forecast, gru_forecast, lstm_forecast) in enumerate(zip(rnn_predictions, gru_predictions, lstm_predictions)):
    predicted_full_rnn[i:i+4] += rnn_forecast
    predicted_full_gru[i:i+4] += gru_forecast
    predicted_full_lstm[i:i+4] += lstm_forecast
    overlap_count[i:i+4] += 1

# Average overlapping predictions
predicted_full_rnn /= overlap_count
predicted_full_gru /= overlap_count
predicted_full_lstm /= overlap_count

# Plot
plt.figure(figsize=(12, 8))

# Plot ground truth as a single continuous line
plt.plot(time_axis_truth, ground_truth_continuous, label="Ground Truth", linestyle="-", color="black")

# Plot the averaged predictions for RNN, GRU, and LSTM
time_axis_predictions = np.arange(time_series_length)
plt.plot(time_axis_predictions, predicted_full_rnn, label="RNN Predictions", linestyle="-", color=params_rnn['rnn']['color'])
plt.plot(time_axis_predictions, predicted_full_gru, label="GRU Predictions", linestyle="--", color=params_gru['gru']['color'])
plt.plot(time_axis_predictions, predicted_full_lstm, label="LSTM Predictions", linestyle=(0, (5, 10)), color=params_lstm['lstm']['color'])

# Finalize the plot
plt.title("RNN, GRU, and LSTM Predictions vs Ground Truth")
plt.xlabel("Time")
plt.ylabel("Yield", fontsize=18)
plt.legend()
plt.grid()
plt.show()
```

```python
def plot_predictions(dataset_, df_period, ax, title):
    params_rnn, rnn_predictions, ground_truth = load_and_predict(dataset_, 'rnn', df_period)
    params_gru, gru_predictions, ground_truth = load_and_predict(dataset_, 'gru', df_period)
    params_lstm, lstm_predictions, ground_truth = load_and_predict(dataset_, 'lstm', df_period)

    # Convert ground truth into a single continuous array
    ground_truth_continuous = []
    for i in range(len(ground_truth) - 1):
        ground_truth_continuous.append(ground_truth[i, 0])
    ground_truth_continuous.extend(ground_truth[-1])
    ground_truth_continuous = np.array(ground_truth_continuous)

    # Time axis for ground truth
    time_axis_truth = np.arange(len(ground_truth_continuous))

    # Initialize arrays for averaged predictions
    time_series_length = len(rnn_predictions) + 3  # Include overlapping days
    predicted_full_rnn = np.zeros(time_series_length)
    predicted_full_gru = np.zeros(time_series_length)
    predicted_full_lstm = np.zeros(time_series_length)
    overlap_count = np.zeros(time_series_length)  # To average overlapping predictions

    # Fill the arrays for predictions (for RNN, GRU, and LSTM)
    for i, (rnn_forecast, gru_forecast, lstm_forecast) in enumerate(zip(rnn_predictions, gru_predictions, lstm_predictions)):
        predicted_full_rnn[i:i+4] += rnn_forecast
        predicted_full_gru[i:i+4] += gru_forecast
        predicted_full_lstm[i:i+4] += lstm_forecast
        overlap_count[i:i+4] += 1

    # Average overlapping predictions
    predicted_full_rnn /= overlap_count
    predicted_full_gru /= overlap_count
    predicted_full_lstm /= overlap_count

    # Time axis for predictions
    time_axis_predictions = np.arange(time_series_length)

    # Plot on the provided axis
    ax.plot(time_axis_truth, ground_truth_continuous, label="Ground Truth", linestyle="-", color="black")
    ax.plot(time_axis_predictions, predicted_full_rnn, label="RNN Predictions", linestyle="-", color='blue')
    ax.plot(time_axis_predictions, predicted_full_gru, label="GRU Predictions", linestyle="--", color=params_gru['gru']['color'])
    ax.plot(time_axis_predictions, predicted_full_lstm, label="LSTM Predictions", linestyle=(0, (5, 10)), color=params_lstm['lstm']['color'])

    # Set title and labels for the subplot
    ax.set_title(title)
    ax.set_xlabel("Time")
    ax.set_ylabel("Yield")
    ax.legend()
    ax.grid()

# Data configurations
datasets = ["2yr", "10yr"]
periods = ["H1_2020", "H1_2023"]
dataframes = {
    ("2yr", "H1_2020"): df_2yr_H1_2020,
    ("2yr", "H1_2023"): df_2yr_H1_2023,
    ("10yr", "H1_2020"): df_10yr_H1_2020,
    ("10yr", "H1_2023"): df_10yr_H1_2023
}

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)

# Loop through datasets and periods to populate the grid
for col, dataset_ in enumerate(datasets):
    for row, period in enumerate(periods):
        df_period = dataframes[(dataset_, period)]
        ax = axes[row, col]
        title = f"{dataset_.replace('yr', '-year UST yields in')} {period.replace('_', ' ')}"
        plot_predictions(dataset_, df_period, ax, title)

# Set overall figure labels
fig.suptitle("RNN, GRU, and LSTM Predictions vs Ground Truth (2x2 Panels)", fontsize=20)
fig.supxlabel("Time", fontsize=24)
fig.supylabel("Yield", fontsize=24)
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the figure
plt.show()

plt.savefig("figures/MOVE_model_performance.png", dpi=1000)
```
